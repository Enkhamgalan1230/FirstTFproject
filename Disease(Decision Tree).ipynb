```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
# spliting dataset into attributes and labels

X_train = dfTrain.iloc[:, :-1].values 
Y_train = dfTrain.iloc[:, 132].values 
X_test = dfTest.iloc[:, :-1].values 
Y_test = dfTest.iloc[:, 132].values 
```

Gini impurity measures how often a randomly chosen element of a set would be incorrectly labeled if it were labeled randomly and independently according to the distribution of labels in the set. It reaches its minimum (zero) when all cases in the node fall into a single target category.

```python
classifierDT = DecisionTreeClassifier(splitter='best', criterion='entropy', min_samples_leaf=2)
classifierDT.fit(X_train, Y_train)
```

```python
Y_predDT = classifierDT.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, Y_predDT))
print(classification_report(Y_test, Y_predDT))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Train Accuracy: ", accuracy_score(Y_train, classifierDT.predict(X_train)))
print("Test Accuracy: ", accuracy_score(Y_test, Y_predDT))

#The dataset is very clean and simple so accuracy is high at this one.
```

```python
imp = classifierDT.feature_importances_
imp
```

```python
columns = list(dfTrain.columns)
columns
```

```python
columns = columns[:132]
column_names = ['symptom', 'importance']
dfImp = np.vstack((columns, imp)).T
dfImp = pd.DataFrame(dfImp, columns = column_names)
dfImp
```

```python
# Increase the figure size
importance_df = pd.DataFrame({'Feature': columns, 'Importance': imp})
importance_df = importance_df.sort_values(by='Importance', ascending=True)

# Set a minimum threshold for feature importance

importance_threshold = np.quantile(importance_df['Importance'], q=0.75)

plt.figure(figsize=(10, len(importance_df) * 0.3))  # Adjust the height based on the number of features

# Create the bar plot with adjusted bar colors
bar_colors = ['red' if x > importance_threshold else 'steelblue' for x in importance_df['Importance']]
sns.barplot(x='Importance', y='Feature', data=importance_df, palette=bar_colors)

# Add the threshold line and annotation
plt.axvline(x=importance_threshold, color='red', linestyle='--')
plt.text(importance_threshold + 0.01, len(importance_df) - (len(importance_df) * 0.1), 'Threshold', color='red',
         rotation=270, verticalalignment='top')  # Adjust the text position and rotation

# Adjust plot details for clarity
plt.title('Feature Importance For Decision Tree Model')
plt.xlabel('Importance Value')
plt.ylabel('Feature')
plt.tight_layout()  # Adjust layout

plt.show()
```

```python
columns_to_drop = importance_df[importance_df['Importance'] < importance_threshold]['Feature']

# Drop these columns from your dataset
dfTrainDrop = dfTrain.drop(columns_to_drop, axis=1)
dfTestDrop = dfTest.drop(columns_to_drop, axis=1)


# Show the first few rows of the new dataset to verify columns are dropped
dfTrainDrop.head()
```

In information theory, the entropy of a random variable is the average level of “information”, “surprise”, or “uncertainty” inherent to the variable's possible outcomes. In the context of Decision Trees, entropy is a measure of disorder or impurity in a node.

```python
dfTrainDrop.info()
```

```python
x_train = dfTrainDrop.iloc[:, :-1].values
y_train = dfTrainDrop.iloc[:, 33].values 
x_test = dfTestDrop.iloc[:, :-1].values 
y_test = dfTestDrop.iloc[:, 33].values

# using DT based on information gain
classifierDT = DecisionTreeClassifier(splitter='best', criterion='entropy', min_samples_leaf=2)
classifierDT.fit(x_train, y_train)
```

```python
y_predDT = classifierDT.predict(x_test)

# using confusion matrix
print(confusion_matrix(y_test, y_predDT))
print(classification_report(y_test, y_predDT))

# using accuracy performance metric
print("Train Accuracy: ", accuracy_score(y_train, classifierDT.predict(x_train)))
print("Test Accuracy: ", accuracy_score(y_test, y_predDT))
```
